Traçage du  graphique de disperssion en utilisant la colonne "area" en abscisse et "price" en ordonnée

```python
X = df["area"]
Y = df["price"]
plt.figure(figsize=(10, 6))
plt.scatter(X,Y, alpha=0.6)
plt.xlabel("Superficie d'une maison")
plt.ylabel("Le prix d'une maison")
plt.title("Graphique montrant le prix d'une maison en fonction de sa superficie")
plt.show()
```

## Analyse

En observant ce graphique, on constate que plus la sperficie d'une maison augmente, plus le prix augmente. De plus, les maisons qui sont à une supericie entre 3000 m² et 5000 m² sont comprisent entre 3 millions à 5 millions environs; les maisons qui sont comprisent entre 6 Millions et 8 Millions ont la plupart une superficie de 6000 m². Enfin, on constate egalement que a partir de 12000 m² et au dessus de 12 Millions, les valeurs sont tres eloigner des autres, ils sont probablement des valeurs aberantes
